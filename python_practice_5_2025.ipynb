5. (9б). Створіть клас 3D-фігура без атрибутів, з методом get_volume для отримання об'єму фігури, що повертає 0 та __add __, який приймає self та other в якості аргументів, а повертає суму об'ємів 3d-фігур self та other.

```python
class Shape3D:
    def get_volume(self):
        return 0

    def __add__(self, other):
        return self.get_volume() + other.get_volume()

shape1 = Shape3D()
shape2 = Shape3D()

print(shape1.get_volume())
print(shape2.get_volume())
print(shape1 + shape2)
```

6. (11б). Створіть 2 дочірніх класи від 3d-фігура: Циліндр та Куб, які мають атрибути, необхідні для розрахунку об'єму. Визначте метод get_volume в кожному з них так, щоби вони розраховували об'єм у залежності від формули для кожного типу фігури. Створіть обʼєкт класу Циліндр та обʼєкт класу Куб. Виконайте операцію суми за допомогою оператора + між ними, аби знайти сумарний об'єм.

```python
import math
class Cylinder(Shape3D):
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height

    def get_volume(self):
        return math.pi * (self.radius ** 2) * self.height


class Cube(Shape3D):
    def __init__(self, side):
        self.side = side

    def get_volume(self):
        return self.side ** 3

cylinder = Cylinder(radius=3, height=10)
cube = Cube(side=4)

print("Cylinder volume:", cylinder.get_volume())
print("Cube volume:", cube.get_volume())

print("Total volume:", cylinder + cube)
```

8. (13б). Створіть клас EWallet з приватними атрибутами balance та id. Реалізуйте методи поповнення та зняття коштів, забезпечивши належну інкапсуляцію. Підказка: використовуйте декоратори property та setter.

```python
class EWallet:
    def __init__(self, wallet_id, initial_balance=0):
        self.__id = wallet_id
        self.__balance = initial_balance

    @property
    def id(self):
        return self.__id

    @property
    def balance(self):
        return self.__balance

    @balance.setter
    def balance(self, new_balance):
        if new_balance < 0:
            print("Помилка: Баланс не може бути від'ємним!")
        else:
            self.__balance = new_balance

    def deposit(self, amount):
        if amount <= 0:
            print("Сума поповнення має бути додатною.")
        else:
            self.__balance += amount
            print(f"Рахунок {self.__id}: поповнено на {amount}. Баланс: {self.__balance}")

    def withdraw(self, amount):
        if amount <= 0:
            print("Сума зняття має бути додатною.")
            return

        if amount > self.__balance:
            print(f"Рахунок {self.__id}: недостатньо коштів для зняття {amount}.")
        else:
            self.__balance -= amount
            print(f"Рахунок {self.__id}: знято {amount}. Баланс: {self.__balance}")

my_wallet = EWallet(wallet_id="UA123456", initial_balance=100)

print(f"ID гаманця: {my_wallet.id}")
print(f"Початковий баланс: {my_wallet.balance}")

my_wallet.deposit(50)
my_wallet.withdraw(30)
my_wallet.withdraw(200)

my_wallet.balance = 500
print(f"Баланс після прямої зміни: {my_wallet.balance}")

my_wallet.balance = -50
```

9. (11б). Створіть клас ShoppingCart, який містить список об'єктів типу Item. Реалізуйте методи для додавання айтемів, їх видалення та відображення списку айтемів.

```python
class Item:
    __counter = 0

    def __init__(self, name):
        Item.__counter += 1
        self.__id = Item.__counter
        self.name = name

    @property
    def id(self):
        return self.__id

    def __str__(self):
        return f"Item {self.id}: {self.name}"


class ShoppingCart:
    def __init__(self):
        self.list_items = []

    def add_item(self, item):
        self.list_items.append(item)

    def delete_item(self, item):
        if item in self.list_items:
            self.list_items.remove(item)

    def show_items(self):
        if not self.list_items:
            print("Cart is empty")
        else:
            for item in self.list_items:
                print(item)

cart = ShoppingCart()

item1 = Item("Pizza")
item2 = Item("Pineapple")
item3 = Item("Cheese")

cart.add_item(item1)
cart.add_item(item2)
cart.add_item(item3)

cart.show_items()

cart.delete_item(item2)
cart.show_items()
```

```python
class Vehicle:
    def __init__(self, brand, production_age):
        self.brand = brand
        self.production_age = production_age


class Engine:
    def __init__(self, engine_type, size):
        self.engine_type = engine_type
        self.size = size


class Car(Vehicle, Engine):
    def __init__(self, brand, production_age, engine_type, size):
        Vehicle.__init__(self, brand, production_age)
        Engine.__init__(self, engine_type, size)

    def __str__(self):
        return (f"Car: {self.brand}, age: {self.production_age} years, "
                f"engine: {self.engine_type}, {self.size} L")

my_car = Car("Toyota", 5, "petrol", 2.0)
print(my_car)
print(Car.mro())
```

 12. (5б). Опишіть своїми словами в одному реченні, як ви розумієте різницю між ітерабельними обʼєктами та ітераторами (можна на прикладі).

Ітерабельний об'єкт це той, по якому можна пройтися циклом (наприклад список), а ітератор - це спеціальний об'єкт, який пам'ятає поточну позицію (наступний елемент після відданого) та повертає елементи послідовно через функцію next

13. (11б). Створіть клас, що має визначений Iterator Protocol та при кожному виклику next повертає назву дня тижня, поки вони не вичерпаються. Зверніть увагу, що порядок має значення.

```python
class WeekDays:
    def __init__(self):
        self.days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index < len(self.days):
            day = self.days[self.index]
            self.index += 1
            return day
        else:
            raise StopIteration

week = WeekDays()
iterator = iter(week)

for day in iterator:
    print(day)
```
